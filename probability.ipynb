```python
# Multiples of 5 
from itertools import product as prod 

n_dice = 5

faces = {1,2,3,4,5,6}

event_space = set(prod(faces,repeat=n_dice))

print(event_space)
```

```python
len(event_space)
```

```python
fav_ot = []

for outcome in event_space:
    d1,d2,d3,d4,d5 = outcome
    if (d1+d2+d3+d4+d5) % 5 == 0 and  (d1+d2+d3+d4+d5) % 3 != 0 :
        fav_ot.append(outcome)
    set(fav_ot)

print(len(fav_ot))

print(f"Probability of the outcome is : {((len(fav_ot) / len(event_space)) * 100):.2f}%")
```
